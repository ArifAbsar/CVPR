```python
import numpy as np
import cv2
from skimage.color import rgb2gray
from fcmeans import FCM
import matplotlib.pyplot as plt

```

```python
def apply_fcm(image_path, n_clusters=20):
    image = cv2.imread(image_path)
    gray_image = rgb2gray(image)
    pixels = gray_image.reshape((-1, 1))
    fcm = FCM(n_clusters=n_clusters)
    fcm.fit(pixels)
    fcm_labels = fcm.u.argmax(axis=1)
    fcm_segmented = fcm_labels.reshape(gray_image.shape)
    return fcm_segmented
```

```python
image_path = 'F:/Machine learning/no_stop.JPG'
fcm_segmented = apply_fcm(image_path)

# Save the segmented image
cv2.imwrite('F:\Machine learning\stop.jpg', fcm_segmented * 255 / fcm_segmented.max())

# Display the original and segmented images
original_image = cv2.imread(image_path)
segmented_image = cv2.imread('F:\Machine learning\stop.jpg', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Segmented Image')
plt.imshow(segmented_image, cmap='gray')
plt.axis('off')

plt.show()
```
